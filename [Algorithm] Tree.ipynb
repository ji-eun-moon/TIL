````python
V = int(input())  #  
arr = list(map(int, input().split()))
E = V - 1  #  
left = [0]*(V+1)  #     
right = [0]*(V+1)  #     
par = [0]*(V+1)  #    
def preorder(i):
    if i:  #  
            print(i, end = ' ')
            preorder(left[i])
            preorder(right[i])
    return

for i in range(E):
    p, c = arr[i*2], arr[i*2 + 1]
    if left[p] == 0:
        left[p] = c
    else:
        right[p] = c
    par[c] = p
root = 1
while par[root] != 0:  # root 
    root += 1

print(root)
preorder(root)
````

````python
arr = ' ABCDEFG'

#  
def preorder(now):
    if now > len(arr)-1 :
        return
    print(arr[now], end=' ')
    preorder(now*2)
    preorder(now*2 + 1)

preorder(1)  # A B D E C F G
print()

#  
def postorder(now):
    if now > len(arr)-1 :
        return
    postorder(now*2)
    postorder(now*2 + 1)
    print(arr[now], end=' ')

postorder(1)  # D E B F G C A
print()

#  
def inorder(now):
    if now > len(arr)-1 :
        return
    inorder(now*2)
    print(arr[now], end=' ')
    inorder(now*2 + 1)

inorder(1)  # D B E A F C G 
print()
````
